## Linear Regression Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

```python
train_x, test_x, train_y, test_y = train_test_split(fin_df['Percent of adults with less than a high school diploma, 2015-19'],
                                                    fin_df['Age-Adjusted Death Rate'],
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
train_x = train_x.values.reshape(-1, 1)
test_x = test_x.values.reshape(-1, 1)
```

```python
print("The first five samples {}".format(train_x[:5]))
print("The first five targets {}".format(train_y[:5]))
print("The number of samples in train set is {}".format(train_x.shape[0]))
print("The number of samples in test set is {}".format(test_x.shape[0]))

lr = LinearRegression()
lr.fit(train_x, train_y)

pred_y = lr.predict(test_x)
print("The first five prediction {}".format(pred_y[:5]))
print("The real first five labels {}".format(test_y[:5]))

mse = metrics.mean_squared_error(test_y, pred_y)
print("Mean Squared Error {}".format(mse))
```

```python
Rsquared = metrics.r2_score(test_y, pred_y)
print('R2 score is {}'.format(Rsquared))
```

```python
# Print the Intercept:
print('intercept:', lr.intercept_)

# Print the Slope:
print('slope:', lr.coef_) 
```

```python
xfit = np.linspace(0, 40, 1000)
yfit = lr.predict(xfit[:, np.newaxis])

plt.scatter(train_x, train_y, s =0.1)
plt.plot(xfit, yfit, c = 'r')
plt.show()
```
